```python
import requests
import pandas as pd

API_KEY = "23fa597d11274e39a2badccc6efe7b15"

def get_news(company_name):
    url = f"https://newsapi.org/v2/everything?q={company_name}+stock&language=en&sortBy=publishedAt&apiKey={API_KEY}"
    response = requests.get(url)
    articles = response.json()['articles']
    
    news_list = []
    for article in articles[:20]:
        news_list.append({
            'date': article['publishedAt'][:10],
            'headline': article['title'],
            'source': article['source']['name']
        })
    
    return pd.DataFrame(news_list)

# Test it
reliance_news = get_news("Reliance Industries")
print(reliance_news)
reliance_news.to_csv("../data/reliance_news.csv", index=False)
```
