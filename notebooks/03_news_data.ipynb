```python
import requests
import pandas as pd
import time

API_KEY = "23fa597d11274e39a2badccc6efe7b15"

companies = ["Reliance Industries", "TCS", "Infosys", "HDFC Bank", "Wipro"]

all_news = []

for company in companies:
    url = f"https://newsapi.org/v2/everything?q={company}+stock&language=en&sortBy=publishedAt&pageSize=100&apiKey={API_KEY}"
    response = requests.get(url)
    articles = response.json().get('articles', [])
    
    for article in articles:
        all_news.append({
            'date': article['publishedAt'][:10],
            'headline': article['title'],
            'company': company
        })
    
    time.sleep(1)  # avoid rate limiting

df = pd.DataFrame(all_news)
print(f"Total headlines collected: {len(df)}")
print(df['date'].value_counts().head(10))
df.to_csv("../data/all_news.csv", index=False)
```
