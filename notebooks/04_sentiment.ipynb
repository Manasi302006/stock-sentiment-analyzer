```python
import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

news = pd.read_csv("../data/reliance_news.csv")

analyzer = SentimentIntensityAnalyzer()

def get_sentiment(text):
    score = analyzer.polarity_scores(str(text))
    compound = score['compound']
    if compound >= 0.05:
        return compound, 'POSITIVE'
    elif compound <= -0.05:
        return compound, 'NEGATIVE'
    else:
        return compound, 'NEUTRAL'

news[['score', 'sentiment']] = news['headline'].apply(
    lambda x: pd.Series(get_sentiment(x))
)

print(news[['headline', 'score', 'sentiment']])
print("\n--- Sentiment Breakdown ---")
print(news['sentiment'].value_counts())
```
